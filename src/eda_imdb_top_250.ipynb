### 1. Descripción de los datos
Esta base de datos contiene la información básica de las 250 peliculas mejor puntuadas en el portal imdb.com

|Campo          |Tipo       |Descripción                            |Notas   |   |
|---------------|-----------|---------------------------------------|---|---|
|rank           |entero     |Posición de la película en el TOP250   |Valor 1 a 250   |   |
|name           |texto      |Nombre de la película                  |Nombre en idioma original   |   |
|year           |entero     |Año de estreno                         |   |   |
|rating         |decimal    |Calificación                           |Otorgado por usuarios   |   |
|genre          |texto      |Generos                                |Lista con los generos a los que pertenece   |   |
|certificate    |texto      |Clasificación                          |Clasificación asignado en el pais de origen   |   |
|run_time       |texto      |Duración                               |Medido en horas y minutos    |   |
|tagline        |texto      |Eslogan                                |En Ingles   |   |
|budget         |entero     |Presupuesto                            |En dolares   |   |

|directors      |texto      |Directores                             |Lista de los directores   |   |
|writers        |texto      |Escritores                             |Lista de los escritores   |   |

La pregunta que quiere responder esta relacionada a la ganancia, pero dado que los valores proporcionados no estan actualizados a la inflación actual, utilizaremos un valor agregado calculando la proporción entre el box_office respecto al presupuesto en valores originales. Este campo lo llamaremos ganancia_proporcional.

$$ ganancia proporcional = {boxoffice \over budget} $$

### 3. Valores faltantes

```python
data.isnull().sum()
```

No contamos con valores faltantes, sin embargo en las columnas de budget y box_office tenemos valores nulos ("Not Available"). Para el ejercicio esta información es indispensable así que eliminamos los registros que no sean numericos.

```python
dataC = data[data['budget'].str.contains("Not Available")==False]
dataC
```

Eliminamos los registros que tienen "Not Available" en la columna budget. Quedando 211 registros

```python
dataD = dataC[dataC['box_office'].str.contains("Not Available")==False]
dataD
```

Eliminamos los registros que tienen "Not Available" en la columna box_office. Quedando 211 registros

```python
dataD.dtypes
```

Tanto budget como box_office deben ser numericos asi que cambiaremos el tipo. Pero antes tenemos que elimnar caracteres que no sean númericos.

```python
dataD['budget'] = dataD['budget'].str.replace('\D', '', regex=True)
dataD['box_office'] = dataD['box_office'].str.replace('\D', '', regex=True)
dataD
```

```python
dataD['budget'] = pd.to_numeric(dataD['budget'])
dataD['box_office'] = pd.to_numeric(dataD['box_office'])
print(dataD)
print(dataD.dtypes)
```

Ya con los datos correctos y con la información un poco más limpia, creamos la columna ganancia proporcional.

Esta ganancia nos permitira comparar los ingresos de las peliculas en base a cada dolar generado en su semana de apertura contra el presupuesto.

```python
dataD['ganancia_proporcional'] = ''
dataD['ganancia_proporcional'] = dataD['box_office']/dataD['budget']
dataD
```

Ya con la información limpia la guardamos en un nuevo dataset y lo almacenamos en la carpeta de results.

```python
dataD.to_csv(r'..\results\IMDB_Top_250_Clean.csv')
```

### Graficas Fallidas

En esta sección intente encontrar una relación entre los valores económicos y los ratings. Lamentablemente la gran diferencia entre los valores mas bajos y los mas altos hacen que la visualización sea muy dificil o irrelevante.

Lo que si se puede observar es como a lo largo de los años el incremento tanto de los presupuestos como de las ganancias ha sido mas o menos constante. La conclusión preliminar es que a través de los años los presupuestos se han incrementado ya sea por la inflación o por una escalada en la robustes del negocio cinematografico.


```python
alt.Chart(dataMovies).mark_rect().encode(
    alt.X("budget:Q", title = "Budget", axis = alt.Axis(labelAngle = 0), scale=alt.Scale(type='log')),
    alt.Y('year:O', title = "Box Office", sort='descending'),
    alt.Color('rating:O', title = "Rating", scale=alt.Scale(scheme='greenblue'))
).properties(title={
    "text": ["Ratings vs $$"], 
    "subtitle": ["como se afectan los raitings de acuerdo al dinero invertido o ingresado"],
    "color": "black",
    "subtitleColor": "grey"},
             width=800, height=400)

```

### Conclusiones

Para poder llegar a una conclusión certera es necesario ampliar la base con mas peliculas de manera que se vean reflejados los comportamientos de manera mas clara y poder determinar si la conclusión preeliminar a la que se llego es correcta.

Además, para una correcta comparación el método de "ganancia proporcional" propuesto no es suficiente, así que la propuesta es trabajar con los presupuestos y homologarlos con base a la inflación.
